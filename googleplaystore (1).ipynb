```python
import pandas as pd
import numpy as np
df = pd.read_csv("googleplaystore.csv");
df.head()
```

```python
#Rating
print(df['Rating'].isnull().sum())
mean_rating = df['Rating'].mean()
df['Rating'].fillna(mean_rating, inplace=True)
df.head()
```

```python
df = df[(df['Rating'] >= 0) & (df['Rating'] <= 5)]
```

```python
df['Rating'] = df['Rating'].astype(str)
df['Rating'] = df['Rating'].str.replace('%', '', regex=True).astype(float)
```

```python
df.drop_duplicates(subset=['App', 'Rating'], keep='first', inplace=True)
```

```python
df['Rating'].describe()
```

```python
df.head()
```

```python
# size
df['Size'] = df['Size'].astype(str)
```

```python
df['Size'] = df['Size'].replace('Varies with device', np.nan)
```

```python
extracted = df['Size'].str.extract(r'(\d+\.?\d*)([MkG]?)')
df['Size_Value'] = extracted[0] 
df['Size_Unit'] = extracted[1]   
```

```python
def convert_to_mb(row):
    value = float(row['Size_Value']) if not pd.isnull(row['Size_Value']) else np.nan
    unit = row['Size_Unit']
    
    if unit == 'G':
        return value * 1024
    elif unit == 'k':
        return value / 1024  
    elif unit == 'M':
        return value
    else:
        return np.nan  

df['Size_MB'] = df.apply(convert_to_mb, axis=1)
```

```python
mean_size = df['Size_MB'].mean()
df['Size_MB'].fillna(mean_size, inplace=True)
```

```python
df['Size_MB'].describe()
```

```python
df['Size_MB'].unique()
```

```python
# price
df['Price'] = df['Price'].astype(str)
df['Price'] = df['Price'].str.replace('$', '', regex=False) 
df['Price'] = df['Price'].str.replace(',', '', regex=False)  
```

```python
df.loc[df['Type'] == 'Free', 'Price'] = '0'
df.loc[(df['Price'] == '0') & (df['Type'] == 'Paid'), 'Type'] = 'Free'
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')  
```

```python
max_price = 200
df = df[df['Price'] <= max_price]
df['Price'] = df['Price'].clip(upper=max_price)

df['Price'].describe()
```

```python
# Android Version
df['Android Ver'].unique()
```

```python
df['Android Ver'].value_counts()
```

```python
df['Android Ver'] = df['Android Ver'].str.replace('Android', '', regex=False).str.strip()

df[['Min_Android_Ver', 'Max_Android_Ver']] = df['Android Ver'].str.split('-|and up|–', expand=True).iloc[:, :2]

df['Min_Android_Ver'] = df['Min_Android_Ver'].str.strip()
df['Max_Android_Ver'] = df['Max_Android_Ver'].str.strip()
```

```python
common_version = df['Min_Android_Ver'].mode()[0]
df['Min_Android_Ver'] = df['Min_Android_Ver'].replace('Varies with device', common_version)
df['Max_Android_Ver'] = df['Max_Android_Ver'].replace('Varies with device', common_version)
```

```python
df['Max_Android_Ver'] = df['Max_Android_Ver'].fillna(df['Min_Android_Ver'])
```

```python
df['Min_Android_Ver'] = pd.to_numeric(df['Min_Android_Ver'], errors='coerce')
df['Max_Android_Ver'] = pd.to_numeric(df['Max_Android_Ver'], errors='coerce')
```

```python
mean_min = df['Min_Android_Ver'].mean()
df['Min_Android_Ver'] = df['Min_Android_Ver'].fillna(mean_min)
df['Max_Android_Ver'] = df['Max_Android_Ver'].fillna(mean_min)
```

```python
df['Current Ver'] = df['Current Ver'].replace('Varies with device', np.nan)
```

```python
latest_version = df['Current Ver'].max()
df['Current Ver'] = df.apply(
    lambda row: latest_version 
    if pd.isnull(row['Current Ver']) else row['Current Ver'],
    axis=1) 
```

```python
# Outlier Detection and Handling
Q1 = df['Price'].quantile(0.25)
Q3 = df['Price'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['Price'] < lower_bound) | (df['Price'] > upper_bound)]
```

```python
median_price = df['Price'].median()
df['Price'] = np.where(df['Price'] > upper_bound, median_price, df['Price'])
```

```python
df
```

```python
df['Price'] = df['Price'].replace('[\$,]', '', regex=True).astype(float)
```

```python
df['Size'] = df['Size'].replace({'Varies with device': np.nan, 'k': '*1e3', 'M': '*1e6'}, regex=True).map(
    lambda x: eval(x) 
    if isinstance(x, str) 
    else x)
```

```python
df['Installs'] = df['Installs'].replace('[+,]', '', regex=True).astype(float)
```

```python
df['Reviews'] = pd.to_numeric(df['Reviews'], errors='coerce')
```

```python
df.to_csv('cleaned_googleplaystore.csv', index=False, encoding='utf-8-sig')
```
